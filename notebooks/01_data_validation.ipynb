```python
from pathlib import Path
import pandas as pd

ruta_datos = Path("../Data/retail_sales_dataset.csv")
if not ruta_datos.exists():
    ruta_datos = Path("Data/retail_sales_dataset.csv")

# Cargar el dataset

df = pd.read_csv(ruta_datos)

# Respuestas solicitadas
precio_maximo = df["Precio Unitario"].max()
mayor_total_ventas = df["Total Venta"].max()

print(f"Precio máximo de producto: {precio_maximo}")
print(f"Mayor total de ventas registrado: {mayor_total_ventas}")
```
